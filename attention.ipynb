```python
import numpy as np
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=1, keepdims=True)
```

```python
def scaled_dot_product_attention(Q, K, V):
    score = np.dot(Q, K.T)
    d_k = K.ndim
    scaled_score = score / np.sqrt(d_k)
    weights = softmax(scaled_score)
    output = np.dot(weights, V)

    return output, weights
```

```python
Q = np.array([[1, 0, 1]])  # shape: (1, d_k)
K = np.array([[1, 0, 0],    # shape: (n, d_k)
              [0, 1, 0],
              [0, 0, 1]])
V = np.array([[1, 10],      # shape: (n, d_v)
              [2, 20],
              [3, 30]])
```

```python
# Attention
output, weights = scaled_dot_product_attention(Q, K, V)
```

```python
print("Output:")
print(output)
print("Weights:")
print(weights)
```
