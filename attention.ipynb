```python
import numpy as np

np.random.seed(42)

def softmax(x):
    e_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e_x / e_x.sum(axis=-1, keepdims=True)

def scaled_dot_product_attention(Q, K, V):
		score = np.dot(Q, K.T)
		d_k = K.shape[-1]
		scaled_score = score / np.sqrt(d_k)
		weights = softmax(scaled_score)
		output = np.dot(weights, V)

		return output, weights

X = np.random.rand(2, 3)
W_q = np.random.rand(X.shape[-1], X.shape[-1]) # 임의의 Query 가중치 행렬
W_k = np.random.rand(X.shape[-1], X.shape[-1]) # 임의의 Key 가중치 행렬
W_v = np.random.rand(X.shape[-1], X.shape[-1]) # 임의의 Value 가중치 행렬

Q = np.dot(X, W_q)
K = np.dot(X, W_k)
V = np.dot(X, W_v)

# Attention
output, weights = scaled_dot_product_attention(Q, K, V)
print(f"X: {X}")
print(f"Output: {output}")
print(f"Weights: {weights}")

# X: [[0.37454012 0.95071431]]
# Output: [[0.79912485 0.56514398]]
# Weights: [[1.]]
```
